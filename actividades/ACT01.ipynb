# Abstract

This work investigates the relationship between linear algebra over complex
numbers and the description of quantum states using Python (NumPy) and Qiskit in
Google Colab. We apply in a practical way the definitions of determinant, trace,
adjoint matrix and tensor product to the analysis of complex matrices and to
the construction of states of one, two and three qubits.

Python functions are implemented to compute determinants and traces of the
matrices $A$, $B$, $C$ and $D$, and these results are expressed in polar form.
Using the adjoint matrix $M^\dagger$, numerical tests of Hermitian
($M \approx M^\dagger$), unitary ($M^\dagger M \approx I$) and normal
($M^\dagger M \approx M M^\dagger$) matrices are programmed, introducing
tolerances to handle floating–point rounding errors. The results show that none
of the four matrices in the statement can be classified as Hermitian, unitary
or normal, whereas the implemented functions correctly identify reference
matrices $H$, $U$ and $N$ that are, respectively, Hermitian, unitary and
normal.

In the qubit framework, we construct the computational bases of
$\mathbb{C}^2$, $\mathbb{C}^4$ and $\mathbb{C}^8$ via the tensor product and we
explicitly obtain the states $|0\rangle$, $|1\rangle$ and the superpositions
$|+\rangle$, $|-\rangle$, $|i+\rangle$, $|i-\rangle$. From the basis of
$\mathbb{C}^4$ we generate the four Bell states $|\Phi^\pm\rangle$,
$|\Psi^\pm\rangle$, and in $\mathbb{C}^8$ we construct the three–qubit states of
the form $(|000\rangle + e^{i\phi}|111\rangle)/\sqrt{2}$ required in the
assignment. Several of the one–qubit states are represented on the Bloch
sphere using Qiskit, numerically verifying that the obtained vectors are
properly normalised and correspond to pure states.

**Keywords:** complex-valued linear algebra; quantum states; tensor product;
Bell states; Hermitian, unitary and normal matrices; Python programming.

### 1.4.1 Números complejos y forma polar

Un número complejo se puede escribir como $z = a + b\,i$, con $a,b\in\mathbb{R}$. Su módulo es $|z| = \sqrt{a^2 + b^2}$ y su argumento (principal) se denota $\operatorname{Arg}(z)\in(-\pi,\pi]$. La forma polar de $z$ es:
$$
z = r\,e^{i\theta}, \qquad r = |z|,\ \theta = \operatorname{Arg}(z)\,.
$$

Esta representación es especialmente útil para describir determinantes y trazas de matrices complejas, ya que permite separar su magnitud (módulo) y su fase (ángulo). En este Trabajo se expresan en forma polar los determinantes y las trazas de las matrices $A$, $B$, $C$ y $D$ dadas en el enunciado.

### 1.4.2 Matrices hermitianas, unitarias y normales

Sea $M\in\mathbb{C}^{n\times n}$. Se define la traspuesta conjugada de $M$ como
$$
M^\dagger = \overline{M}^T\,,
$$
es decir, la matriz obtenida al transponer $M$ y tomar el conjugado complejo de cada entrada. Con esta notación:

- $M$ es **hermitiana** si $M = M^\dagger$.
- $M$ es **unitaria** si $M^\dagger M = I$ (equivalentemente, $M\,M^\dagger = I$).
- $M$ es **normal** si $M^\dagger M = M\,M^\dagger$.

Obsérvese que las matrices hermitianas y las matrices unitarias son casos particulares de matrices normales. El teorema espectral establece que toda matriz normal es diagonalizable mediante una matriz unitaria. En otras palabras, dado un $M$ normal, existe una matriz unitaria $U$ tal que
$$
M = U\,D\,U^\dagger\,,
$$
donde $D$ es una matriz diagonal formada por los valores propios de $M$ (Horn y Johnson, 2013; Kreyszig, 2011).

En este Trabajo se utilizan tres matrices de prueba $H$, $U$ y $N$ para verificar el correcto funcionamiento del código:
$$
H =
\begin{pmatrix}
2 & i \\
-i & 3
\end{pmatrix},
\qquad
U = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix},
\qquad
N =
\begin{pmatrix}
1 & 0 \\
0 & 2i
\end{pmatrix}\,.
$$

**Hermiticidad de $H$.**  
Calculamos su traspuesta conjugada:
$$
H^\dagger = \overline{H}^T =
\begin{pmatrix}
2 & i \\
-i & 3
\end{pmatrix} = H\,.
$$
Por tanto, $H = H^\dagger$ y $H$ es hermitiana.

**Unitariedad de $U$.**  
Su traspuesta conjugada es
$$
U^\dagger = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & -1 \\
1 & 1
\end{pmatrix}\,.
$$
Entonces,
$$
U^\dagger U =
\frac{1}{2}
\begin{pmatrix}
1 & -1 \\
1 &  1
\end{pmatrix}
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix}
=
\frac{1}{2}
\begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
= I\,,
$$
y análogamente se comprueba que $U\,U^\dagger = I$. Por tanto, $U$ es unitaria.

**Normalidad de $N$.**  
La traspuesta conjugada de $N$ es
$$
N^\dagger =
\begin{pmatrix}
1 & 0 \\
0 & -2i
\end{pmatrix}\,.
$$
Entonces,
$$
N^\dagger N =
\begin{pmatrix}
1 & 0 \\
0 & -2i
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & 2i
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 4
\end{pmatrix},
$$
$$
N\,N^\dagger =
\begin{pmatrix}
1 & 0 \\
0 & 2i
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & -2i
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 4
\end{pmatrix}\,.
$$
Como $N^\dagger N = N\,N^\dagger$, la matriz $N$ es normal. No es hermitiana, porque $N \neq N^\dagger$ (el elemento de la posición $(2,2)$ difiere), ni unitaria, ya que $N^\dagger N = \operatorname{diag}(1,4) \neq I$.

Estas tres matrices se emplean en el trabajo como casos de referencia para validar las funciones de clasificación implementadas en Python.

### 1.4.3 Valores propios, vectores propios y ejemplos espectrales

Dado $M\in\mathbb{C}^{n\times n}$ y un vector no nulo $v\in\mathbb{C}^n$, se dice que $v$ es un **vector propio** de $M$ con valor propio $\lambda$ si
$$
M\,v = \lambda\,v\,.
$$
El conjunto de todos los valores propios de $M$ se denomina espectro de $M$. Cuando $M$ es normal, el teorema espectral garantiza la existencia de una base ortonormal de vectores propios y una descomposición $M = U\,D\,U^\dagger$ con $U$ unitaria y $D$ diagonal (Horn y Johnson, 2013).

Para la matriz hermitiana
$$
H =
\begin{pmatrix}
2 & i \\
-i & 3
\end{pmatrix},
$$
el polinomio característico es
$$
\det(H-\lambda I) = \lambda^2 - 5\lambda + 5\,,
$$
cuyas raíces son
$$
\lambda_{1,2} = \frac{5 \pm \sqrt{5}}{2}\,,
$$
ambas reales, como predice la teoría de matrices hermitianas. Los vectores propios asociados a $\lambda_1$ y $\lambda_2$ pueden elegirse ortogonales entre sí y, tras normalizarlos, forman una base ortonormal de $\mathbb{C}^2$.

Para la matriz unitaria
$$
U = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
-1 & 1
\end{pmatrix},
$$
sus valores propios son
$$
\mu_{1,2} = e^{\pm i\pi/4} =
\frac{\sqrt{2}}{2} \pm i\,\frac{\sqrt{2}}{2}\,,
$$
que verifican $|\mu_1| = |\mu_2| = 1$, propiedad característica de las matrices unitarias. Los vectores propios correspondientes pueden normalizarse de modo que formen una base ortonormal, y la matriz $U$ se puede escribir como $U = V\,D\,V^\dagger$ con $V$ unitaria y $D = \operatorname{diag}(\mu_1,\mu_2)$.

Para la matriz normal
$$
N =
\begin{pmatrix}
1 & 0 \\
0 & 2i
\end{pmatrix},
$$
el polinomio característico es
$$
\det(N-\lambda I) = (1-\lambda)(2i-\lambda)\,,
$$
de donde se obtiene que sus valores propios son
$$
\nu_1 = 1,\qquad \nu_2 = 2i\,.
$$
Un conjunto de vectores propios linealmente independientes es
$$
v_1 =
\begin{pmatrix}
1 \\
0
\end{pmatrix},
\qquad
v_2 =
\begin{pmatrix}
0 \\
1
\end{pmatrix},
$$
que ya forman una base ortonormal de $\mathbb{C}^2$. En este caso, la diagonalización unitaria es inmediata, con $U = I$ y
$$
D = \operatorname{diag}(1,\,2i)\,.
$$

Estos ejemplos ilustran de forma explícita las propiedades espectrales características de matrices hermitianas, unitarias y normales. Dichos resultados teóricos se utilizan como referencia en la actividad y sirven, además, para validar las funciones de clasificación programadas en Python.

### 1.4.4 Espacios de Hilbert, base computacional y producto tensorial

La descripción de sistemas cuánticos compuestos se realiza mediante el producto tensorial de espacios de Hilbert (Nielsen y Chuang, 2010). Para un solo cúbit, el espacio de estados es $\mathbb{C}^2$, con la base computacional definida por
$$
|0\rangle =
\begin{pmatrix}
1 \\
0
\end{pmatrix},
\qquad
|1\rangle =
\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$$

A partir de esta base, podemos construir estados de superposición como
$$
|+\rangle = \frac{1}{\sqrt{2}}\big(|0\rangle + |1\rangle\big),\qquad
|-\rangle = \frac{1}{\sqrt{2}}\big(|0\rangle - |1\rangle\big),
$$
$$
|i+\rangle = \frac{1}{\sqrt{2}}\big(|0\rangle + i\,|1\rangle\big),\qquad
|i-\rangle = \frac{1}{\sqrt{2}}\big(|0\rangle - i\,|1\rangle\big),
$$
que en forma de vectores columna son, respectivamente,
$$
|+\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix} 1 \\ 1 \end{pmatrix},\quad
|-\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix} 1 \\ -1 \end{pmatrix},\quad
|i+\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix} 1 \\ i \end{pmatrix},\quad
|i-\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix} 1 \\ -i \end{pmatrix}.
$$

Para dos cúbits, el espacio de estados es
$$
\mathbb{C}^2 \otimes \mathbb{C}^2 \simeq \mathbb{C}^4.
$$
La base computacional se construye mediante el **producto tensorial** ($\otimes$), que para vectores consiste en multiplicar cada componente del primer vector por todos los componentes del segundo.

Una manera compacta de expresar la definición en este contexto es
$$
|a\rangle\otimes|b\rangle =
\begin{pmatrix} a_1 \\ a_2 \end{pmatrix}
\otimes
\begin{pmatrix} b_1 \\ b_2 \end{pmatrix}
=
\begin{pmatrix}
a_1\,b_1 \\
a_1\,b_2 \\
a_2\,b_1 \\
a_2\,b_2
\end{pmatrix}.
$$
Por ejemplo, para el estado $|1\rangle\otimes|0\rangle$ obtenemos directamente
$$
|1\rangle\otimes|0\rangle =
\begin{pmatrix}0\\1\end{pmatrix}
\otimes
\begin{pmatrix}1\\0\end{pmatrix}
=
\begin{pmatrix}
0\times\begin{pmatrix}1\\0\end{pmatrix}\\[2pt]
1\times\begin{pmatrix}1\\0\end{pmatrix}
\end{pmatrix}
=
\begin{pmatrix}
0\\0\\1\\0
\end{pmatrix},
$$
que coincide con la tercera columna canónica de $\mathbb{C}^4$.

Tomando de nuevo
$$
|0\rangle =
\begin{pmatrix} 1 \\ 0 \end{pmatrix},
\qquad
|1\rangle =
\begin{pmatrix} 0 \\ 1 \end{pmatrix},
$$
obtenemos explícitamente:

- Para $|0\rangle \otimes |0\rangle$:
  $$
  |00\rangle
  = |0\rangle \otimes |0\rangle
  = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
    \otimes
    \begin{pmatrix} 1 \\ 0 \end{pmatrix}
  = \begin{pmatrix}
    1\cdot 1 \\
    1\cdot 0 \\
    0\cdot 1 \\
    0\cdot 0
    \end{pmatrix}
  = \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}.
  $$

- Para $|0\rangle \otimes |1\rangle$:
  $$
  |01\rangle
  = |0\rangle \otimes |1\rangle
  = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
    \otimes
    \begin{pmatrix} 0 \\ 1 \end{pmatrix}
  = \begin{pmatrix}
    1\cdot 0 \\
    1\cdot 1 \\
    0\cdot 0 \\
    0\cdot 1
    \end{pmatrix}
  = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}.
  $$

- Para $|1\rangle \otimes |0\rangle$ (ya visto) tenemos
  $$
  |10\rangle =
  \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix}.
  $$

- Para $|1\rangle \otimes |1\rangle$:
  $$
  |11\rangle
  = |1\rangle \otimes |1\rangle
  = \begin{pmatrix} 0 \\ 1 \end{pmatrix}
    \otimes
    \begin{pmatrix} 0 \\ 1 \end{pmatrix}
  = \begin{pmatrix}
    0\cdot 0 \\
    0\cdot 1 \\
    1\cdot 0 \\
    1\cdot 1
    \end{pmatrix}
  = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}.
  $$

Por lo tanto, la base computacional de $\mathbb{C}^4$ queda
$$
|00\rangle =
\begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix},\quad
|01\rangle =
\begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix},\quad
|10\rangle =
\begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix},\quad
|11\rangle =
\begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}.
$$

En general, para tres cúbits el espacio de estados es
$$
\mathbb{C}^2 \otimes \mathbb{C}^2 \otimes \mathbb{C}^2 \simeq \mathbb{C}^8,
$$
y la base computacional se construye de forma análoga a partir de todos los kets de la forma $|b_1 b_2 b_3\rangle$ con $b_j\in\{0,1\}$. A modo de ejemplo, comenzando con
$$
|00\rangle = |0\rangle\otimes|0\rangle =
\begin{pmatrix}1\\0\\0\\0\end{pmatrix},
$$
se obtiene
$$
|000\rangle
= |00\rangle \otimes |0\rangle
= \begin{pmatrix}1\\0\\0\\0\end{pmatrix}
  \otimes
  \begin{pmatrix}1\\0\end{pmatrix}
=
\begin{pmatrix}
1\\0\\0\\0\\0\\0\\0\\0
\end{pmatrix},
$$
y de forma similar
$$
|111\rangle =
\begin{pmatrix}
0\\0\\0\\0\\0\\0\\0\\1
\end{pmatrix}.
$$

En el Trabajo, estas bases computacionales se construyen tanto de forma teórica como utilizando NumPy (función `np.kron`) y Qiskit (métodos `Statevector.tensor` y `Statevector.from_label`), y sirven como punto de partida para definir los estados de Bell y los estados de tres cúbits solicitados en el enunciado.

### 1.4.5 Qúbits, matrices de Pauli, valores propios y esfera de Bloch

Un cúbit es un sistema cuántico de dos niveles cuyo estado se puede describir como un vector unitario en $\mathbb{C}^2$. En general, un estado puro de cúbit se puede escribir como
$$
|\psi\rangle = \alpha\,|0\rangle + \beta\,|1\rangle,
$$
con $\alpha,\beta\in\mathbb{C}$ y $|\alpha|^2 + |\beta|^2 = 1$.

Las matrices de Pauli se definen como
$$
\sigma_x =
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix},
\qquad
\sigma_y =
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix},
\qquad
\sigma_z =
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}\,.
$$

Los estados $|+\rangle$ y $|-\rangle$ se obtienen como autovectores de $\sigma_x$. El polinomio característico de $\sigma_x$ es
$$
\det(\sigma_x - \lambda I) = \lambda^2 - 1,
$$
de donde se obtiene $\lambda = \pm 1$.

- Para $\lambda = +1$ se obtiene un autovector proporcional a $(1,\,1)^T$, que tras normalizar da
  $$
  |+\rangle = \frac{1}{\sqrt{2}}
  \begin{pmatrix}
  1 \\ 1
  \end{pmatrix}
  = \frac{1}{\sqrt{2}}\big(|0\rangle + |1\rangle\big).
  $$
- Para $\lambda = -1$ se obtiene un autovector proporcional a $(1,\,-1)^T$, que tras normalizar da
  $$
  |-\rangle = \frac{1}{\sqrt{2}}
  \begin{pmatrix}
  1 \\ -1
  \end{pmatrix}
  = \frac{1}{\sqrt{2}}\big(|0\rangle - |1\rangle\big).
  $$

De forma análoga, para $\sigma_y$ el polinomio característico es también $\lambda^2 - 1 = 0$, por lo que los valores propios son $\lambda = \pm 1$. Sus autovectores llevan de forma natural a los estados
$$
|i+\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\ i
\end{pmatrix}
= \frac{1}{\sqrt{2}}\big(|0\rangle + i\,|1\rangle\big),\qquad
|i-\rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\ -i
\end{pmatrix}
= \frac{1}{\sqrt{2}}\big(|0\rangle - i\,|1\rangle\big),
$$
que satisfacen
$$
\sigma_y|i+\rangle = +\,|i+\rangle,\qquad
\sigma_y|i-\rangle = -\,|i-\rangle.
$$

Todos estos estados ($|+\rangle$, $|-\rangle$, $|i+\rangle$, $|i-\rangle$) se emplean más adelante en la actividad y algunos de ellos se representan en la esfera de Bloch.

Todo estado puro de un cúbit puede asociarse a un punto de la esfera unitaria en $\mathbb{R}^3$ mediante el llamado vector de Bloch $(x,y,z)$. Las componentes de este vector vienen dadas por
$$
x = \langle\sigma_x\rangle,\quad
y = \langle\sigma_y\rangle,\quad
z = \langle\sigma_z\rangle\,,
$$
donde $\langle\cdot\rangle$ denota el valor esperado en el estado $|\psi\rangle$ considerado (Nielsen y Chuang, 2010). En este trabajo, dicha representación geométrica se implementa con Qiskit mediante la clase `Bloch` y la función `plot_bloch_multivector`.

### 4.3 Determinantes, trazas y forma polar

Ahora calculamos el determinante y la traza de cada una de las matrices
\(A\), \(B\), \(C\) y \(D\).

Recordamos que:

El determinante de una matriz cuadrada es un número complejo que indica algunas propiedades de la matriz, como la invertibilidad.
La traza de una matriz cuadrada es la suma de sus elementos diagonales.

Dado un número complejo $z = a + bi$, su forma polar se escribe como

$$
z = r e^{i\theta},
$$

donde $r = |z|$ es el módulo de $z$  y  $\theta = \operatorname{Arg}(z)$ es el ángulo que forma $z$ con el eje real
(en radianes). En Python podemos obtener $(r, \theta)$ a partir de $z$ usando
la función `cmath.polar`.


En el siguiente bloque definimos una función auxiliar `det_traza_polar(M, nombre)`
que recibe como argumentos una matriz `M` y una cadena `nombre`. Dentro de la
función se calcula primero el determinante mediante `np.linalg.det(M)` y la traza
con `np.trace(M)`.

Después, usamos la función `polar` del módulo `cmath`. En concreto,
`cmath.polar(z)` recibe un número complejo `z` y devuelve una tupla `(r, theta)`,
Por eso escribimos:

- `r_det, theta_det = cmath.polar(det)` para obtener la forma polar del determinante,  
- `r_tr, theta_tr = cmath.polar(tr)` para obtener la forma polar de la traza.

La función muestra por pantalla, para cada matriz, el valor del determinante
y de la traza en forma cartesiana, junto con su módulo y su argumento (en radianes).
Para mejorar la legibilidad, los módulos y los ángulos se muestran con tres cifras
decimales de precisión mediante el formato `:.3f` dentro de las f-strings*.

Finalmente, se llama a `det_traza_polar` para cada una de las matrices $A$, $B$,
$C$ y $D$, pasando también el nombre correspondiente como cadena de texto,
de modo que la salida quede claramente etiquetada.


```python
np.set_printoptions(precision=4, suppress=True) #elegimos commo salida para todo el programa hasta máximo 4 decimales
def det_traza_polar(M, nombre):
    det = np.linalg.det(M)
    tr = np.trace(M)

    r_det, theta_det = cmath.polar(det)
    r_tr, theta_tr = cmath.polar(tr)

    print(f"Matriz {nombre}")
    print(f"det({nombre}) = {det}")
    print(f"|det({nombre})| = {r_det:.3f}, Arg(det({nombre})) = {theta_det:.3f} rad")
    print()
    print(f"tr({nombre}) = {tr}")
    print(f"|tr({nombre})| = {r_tr:.3f}, Arg(tr({nombre})) = {theta_tr:.3f} rad")
    print("-" * 50) #separador entre las propiedades de las matrices

det_traza_polar(A, "A")
det_traza_polar(B, "B")
det_traza_polar(C, "C")
det_traza_polar(D, "D")
```

### 4.4 Cálculo de la expresión combinada

En esta sección calculamos la expresión

$$
E \;=\; 2i\bigl(\det(D) + \operatorname{tr}(C)\bigr)\,AB
\;-\;
(1+i)\,\det(C)\,\operatorname{tr}(D)\,BA.
$$

Observemos que:

- $\det(C)$, $\det(D)$, $\operatorname{tr}(C)$ y $\operatorname{tr}(D)$ son escalares complejos.

- $AB$ y $BA$ son matrices de tamaño $3\times 3$, ya que $A$ y $B$ son $3\times 3$.

- En Python utilizamos el operador `@` para denotar el producto matricial.

En el código siguiente calculamos primero los escalares
$\det(C)$, $\det(D)$, $\operatorname{tr}(C)$ y $\operatorname{tr}(D)$, y a
continuación construimos los dos términos que se restan en la expresión para obtener finalmente la matriz $E$.


```python
detC = np.linalg.det(C)
detD = np.linalg.det(D)
trC = np.trace(C)
trD = np.trace(D)

term1 = 2j * (detD + trC) * (A @ B)
term2 = (1 + 1j) * detC * trD * (B @ A)

E = term1 - term2

print("Matriz E = 2i(det(D) + tr(C)) AB - (1+i) det(C) tr(D) BA\n")
print(E)
```

### 4.5 Matrices inversas

En esta sección calculamos la inversa de las matrices \(A\), \(B\), \(C\) y \(D\),

Recordamos que una matriz cuadrada \(M\) es **invertible** si y solo si su
determinante es distinto de cero. En programación, cuando trabajamos con

En el código que sigue:

ii) Si $|\det(M)|$ es menor que un cierto umbral, se considera que la matriz no es invertible y se muestra un mensaje indicando que tiene determinante nulo o numéricamente nulo e imprime el valor del determinante en pantalla.

### 4.6 Clasificación de las matrices a partir de la definición

En esta sección clasificamos las matrices $A$, $B$, $C$ y $D$ utilizando
directamente las definiciones de matriz hermitiana, unitaria y normal.

Dada una matriz $M \in \mathbb{C}^{n \times n}$, denotamos por $M^\dagger$ su
traspuesta conjugada, es decir, $M^\dagger = \overline{M}^{\,T}$.

Para cada matriz realizamos las siguientes comprobaciones numéricas:

i) Calculamos la traspuesta conjugada $M^\dagger$ y verificamos si
$M = M^\dagger$. Si se cumple, dentro de un margen de tolerancia, consideramos que la matriz es hermitiana.

ii) Si la matriz no es hermitiana, comprobamos la unitariedad calculando $M^\dagger M$ y $M M^\dagger$ y verificando si ambas son, aproximadamente, iguales a la identidad $I$. Si se cumple, consideramos que la matriz es unitaria.

iii) Sabemos que toda matriz hermitiana o unitaria es normal. Por tanto, si la matriz es hermitiana o unitaria, no es necesario comprobar nada más.

iv) Solo si la matriz no es hermitiana ni unitaria comprobamos la normalidad verificando si se satisface $M^\dagger M \approx M M^\dagger$.

```python
import numpy as np

def conjugada_traspuesta(M):
    r"""Calcula la traspuesta conjugada (M^\dagger)."""
    return M.conj().T

def es_hermitiana(M, tol=1e-10):
    r"""Comprueba si M = M^\dagger."""
    return np.allclose(M, conjugada_traspuesta(M), atol=tol)

def es_unitaria(M, tol=1e-10):
    r"""Comprueba si M^\dagger M = I (Unitaria)."""
    n = M.shape[0]
    I = np.eye(n, dtype=complex)
    Mdag = conjugada_traspuesta(M)
    return np.allclose(Mdag @ M, I, atol=tol)

def es_normal(M, tol=1e-10):
    r"""Comprueba si M^\dagger M = M M^\dagger (Normal)."""
    Mdag = conjugada_traspuesta(M)
    return np.allclose(Mdag @ M, M @ Mdag, atol=tol)

def clasificar_matriz(M, nombre, tol=1e-10):
    r"""
    Ejecuta las tres comprobaciones básicas y muestra el resultado.
    """
    print(f"--- Clasificación de Matriz {nombre} ---")

    es_h = es_hermitiana(M, tol)
    es_u = es_unitaria(M, tol)
    es_n = es_normal(M, tol)

    print(f"¿Hermitiana? {es_h}")
    print(f"¿Unitaria?   {es_u}")
    print(f"¿Normal?     {es_n}")

    print("-" * 50)

clasificar_matriz(A, "A")
clasificar_matriz(B, "B")
clasificar_matriz(C, "C")
clasificar_matriz(D, "D")
```

### 4.7 Valores propios y vectores propios

Como apoyo a la clasificación obtenida en la sección anterior a partir de
las definiciones, calculamos ahora los valores propios  y vectores
propios de las matrices $A$, $B$, $C$ y $D$.

Para ello recordemos que:

i) Si una matriz es hermitiana, sus valores propios son reales y puede elegirse una base ortonormal de autovectores.

ii) Si una matriz es unitaria, todos sus valores propios tienen módulo igual a $1$ y también admite una base ortonormal de autovectores.

iii) En nuestro caso, ya hemos visto a partir de la definición que las matrices consideradas no son hermitianas, ni unitarias, ni normales. Por tanto, esperamos que su espectro no presente estas propiedades especiales. Aun así, calculamos los valores propios y autovectores para comprobar **numéricamente** cuál es su comportamiento.

Para el análisis numérico nos enfocamos en tres aspectos:

i) En la línea de **módulos**:  
si las matrices fueran unitarias, todos los módulos de los valores propios
deberían estar muy cerca de $1$.  

ii) En los **valores propios**:  
si las matrices fueran hermitianas, los valores propios deberían ser
aproximadamente reales.  

iii) En la **matriz de Gram** $V^\dagger V$ (donde las columnas de $V$ son los
autovectores obtenidos):  
si esta matriz fuese casi la identidad, los autovectores formarían una familia
casi ortonormal.

```python
def espectro_matriz(M, nombre, tol=1e-8, decimales=6):
    vals, vecs = np.linalg.eig(M)
    G = vecs.conj().T @ vecs
    n = M.shape[0]
    I = np.eye(n, dtype=complex)

    # Tests con tolerancia
    aprox_reales = np.allclose(vals.imag, 0, atol=tol)
    aprox_mod_uno = np.allclose(np.abs(vals), 1, atol=tol)
    ortonormales = np.allclose(G, I, atol=tol)

    print(f"--- Matriz {nombre} ---\n")

    print("Valores propios:")
    print(np.round(vals, decimales))
    print()

    print("Módulos de los valores propios:")
    print(np.round(np.abs(vals), decimales))
    print()

    print("Matriz de Gram V^† V:")
    print(np.round(G, decimales))
    print()

    print(f"¿Valores propios aprox. reales? {aprox_reales}")
    print(f"¿Módulos aprox.  1?      {aprox_mod_uno}")
    print(f"¿Autovectores aprox. ortonormales (V^†V≈I)?  {ortonormales}")
    print("-" * 50)
```

### 4.8 Prueba del código con matrices de referencia


### 4.10 Estados $|+\rangle$, $|-\rangle$, $|i+\rangle$ y $|i-\rangle$

En la Introducción se ha obtenido de forma analítica la expresión de los estados
$|+\rangle$, $|-\rangle$, $|i+\rangle$ y $|i-\rangle$ como autovectores de las matrices de Pauli $\sigma_x$ y $\sigma_y$. En particular, se ha mostrado que, en la base computacional $\{|0\rangle, |1\rangle\}$, estos estados se pueden escribir como

$$
|+\rangle = \frac{1}{\sqrt{2}}\bigl(|0\rangle + |1\rangle\bigr), \qquad
|-\rangle = \frac{1}{\sqrt{2}}\bigl(|0\rangle - |1\rangle\bigr),
$$

$$
|i+\rangle = \frac{1}{\sqrt{2}}\bigl(|0\rangle + i\,|1\rangle\bigr), \qquad
|i-\rangle = \frac{1}{\sqrt{2}}\bigl(|0\rangle - i\,|1\rangle\bigr).
$$


Estos cuatro estados tienen una interpretación geométrica muy clara en la esfera
de Bloch:

i) $|0\rangle$ y $|1\rangle$ corresponden a los polos norte y sur  de la esfera de Bloch.  
ii) $|+\rangle$ y $|-\rangle$ se sitúan sobre el eje $x$, en las direcciones
positiva y negativa respectivamente.  
iii) $|i+\rangle$ y $|i-\rangle$ se sitúan sobre el eje $y$, en las direcciones
positiva y negativa respectivamente.

En las siguientes celdas construimos estos estados a partir de $|0\rangle$ y
$|1\rangle$ usando NumPy y Qiskit, y los representamos gráficamente en la
esfera de Bloch con la función `plot_bloch_multivector`.


```python
# Construcción de los estados a partir de |0> y |1>
ket_plus  = (ket0 + ket1) / np.sqrt(2)
ket_minus = (ket0 - ket1) / np.sqrt(2)
ket_ip    = (ket0 + 1j * ket1) / np.sqrt(2)
ket_im    = (ket0 - 1j * ket1) / np.sqrt(2)

print("|+>:",  ket_plus)
print("|->:",  ket_minus)
print("|i+>:", ket_ip)
print("|i->:", ket_im)
print("-" * 40)

# Pasamos a Statevector de Qiskit
sv_plus  = Statevector(ket_plus)
sv_minus = Statevector(ket_minus)
sv_ip    = Statevector(ket_ip)
sv_im    = Statevector(ket_im)
```

```python
from qiskit.visualization.bloch import Bloch  # <- cambio importante

# Lista de estados que queremos dibujar
estados = [sv_0, sv_1, sv_plus, sv_minus, sv_ip, sv_im]

b = Bloch()

# Opcional: colores distintos para cada estado (en el mismo orden que 'estados')
b.vector_color = ["b", "r", "g", "m", "c", "y"]

for sv in estados:
    alpha = sv.data[0]
    beta  = sv.data[1]
    x = 2 * np.real(np.conj(alpha) * beta)
    y = 2 * np.imag(np.conj(alpha) * beta)
    z = np.abs(alpha)**2 - np.abs(beta)**2
    b.add_vectors([x, y, z])

b.show()

```

### 4.11 Estados de Bell

En la Introducción se ha obtenido de forma detallada la expresión de los cuatro
estados de Bell a partir de la base computacional y de las condiciones de
ortogonalidad, normalización y entrelazamiento máximo. El resultado es que,
en la base $\{|00\rangle, |01\rangle, |10\rangle, |11\rangle\}$, los estados de
Bell se pueden escribir como

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}}\bigl(|00\rangle + |11\rangle\bigr),\qquad
|\Phi^-\rangle = \frac{1}{\sqrt{2}}\bigl(|00\rangle - |11\rangle\bigr),
$$

$$
|\Psi^+\rangle = \frac{1}{\sqrt{2}}\bigl(|01\rangle + |10\rangle\bigr),\qquad
|\Psi^-\rangle = \frac{1}{\sqrt{2}}\bigl(|01\rangle - |10\rangle\bigr).
$$

En esta sección utilizamos dichas expresiones para construir los estados de Bell
en Python. Primero generamos la base computacional de $\mathbb{C}^4$ mediante
el producto tensorial de $|0\rangle$ y $|1\rangle$, y a partir de ella
obtenemos los cuatro estados de Bell usando NumPy y Qiskit.


```python
ket00 = np.kron(ket0, ket0)
ket01 = np.kron(ket0, ket1)
ket10 = np.kron(ket1, ket0)
ket11 = np.kron(ket1, ket1)

phi_plus  = (ket00 + ket11) / np.sqrt(2)
phi_minus = (ket00 - ket11) / np.sqrt(2)
psi_plus  = (ket01 + ket10) / np.sqrt(2)
psi_minus = (ket01 - ket10) / np.sqrt(2)

sv_phi_plus  = Statevector(phi_plus)
sv_phi_minus = Statevector(phi_minus)
sv_psi_plus  = Statevector(psi_plus)
sv_psi_minus = Statevector(psi_minus)
print("|Φ+> =", sv_phi_plus)
print("|Φ-> =", sv_phi_minus)
print("|Ψ+> =", sv_psi_plus)
print("|Ψ-> =", sv_psi_minus)
```

### 4.12 Base computacional de $\mathbb{C}^4$

En la Introducción se ha construido de forma teórica el producto tensorial de
espacios de Hilbert y se ha visto que, para dos qubits, el espacio de estados es
$$
\mathbb{C}^2 \otimes \mathbb{C}^2 \simeq \mathbb{C}^4.
$$
A partir de la base computacional de $\mathbb{C}^2$, dada por
$\{\,|0\rangle, |1\rangle\,\}$, se obtiene la base computacional de
$\mathbb{C}^4$ mediante productos tensoriales:
$$
|00\rangle = |0\rangle \otimes |0\rangle,\qquad
|01\rangle = |0\rangle \otimes |1\rangle,\qquad
|10\rangle = |1\rangle \otimes |0\rangle,\qquad
|11\rangle = |1\rangle \otimes |1\rangle.
$$

En esta sección implementamos esta construcción en Python de dos maneras
equivalentes:

1. Usando la función `np.kron` de NumPy, que calcula el producto tensorial de
   dos vectores.
2. Usando la clase `Statevector` de Qiskit y su operación `tensor`.

En ambos casos obtenemos los cuatro vectores de la base computacional de
$\mathbb{C}^4$ escritos en la base ordenada
$\{\,|00\rangle, |01\rangle, |10\rangle, |11\rangle\,\}$.


```python

ket00 = np.kron(ket0, ket0)
ket01 = np.kron(ket0, ket1)
ket10 = np.kron(ket1, ket0)
ket11 = np.kron(ket1, ket1)
print("opción con kron de numpy")
print("|00> =", ket00)
print("|01> =", ket01)
print("|10> =", ket10)
print("|11> =", ket11)


# Estados de un solo qubit
sv_0 = Statevector.from_label('0')
sv_1 = Statevector.from_label('1')

# Base computacional de C^4 como productos tensoriales en Qiskit
sv_00 = sv_0.tensor(sv_0)
sv_01 = sv_0.tensor(sv_1)
sv_10 = sv_1.tensor(sv_0)
sv_11 = sv_1.tensor(sv_1)
print("\n Opción con Qizkit")
print("|00> =", sv_00.data)
print("|01> =", sv_01.data)
print("|10> =", sv_10.data)
print("|11> =", sv_11.data)
```

### 4.13 Estados de tres qubits con Qiskit

En la Introducción se ha descrito cómo construir estados de tres qubits en el
espacio
$$
\mathbb{C}^2 \otimes \mathbb{C}^2 \otimes \mathbb{C}^2 \simeq \mathbb{C}^8
$$
a partir de los vectores base $|0\rangle$ y $|1\rangle$ y del producto
tensorial.

En particular, consideramos los estados
$$
|\psi_1\rangle =
\frac{1}{\sqrt{2}}\,|000\rangle + \frac{1}{\sqrt{2}}\,|111\rangle,
$$
$$
|\psi_2\rangle =
\frac{i}{\sqrt{2}}\,|000\rangle - \frac{1}{\sqrt{2}}\,|111\rangle,
$$
$$
|\psi_3\rangle =
\frac{1}{\sqrt{2}}\,|000\rangle + \frac{i}{\sqrt{2}}\,|111\rangle.
$$

En esta sección construimos estos estados directamente con Qiskit usando la
clase `Statevector`. Primero generamos los estados base $|000\rangle$ y
$|111\rangle$ mediante `Statevector.from_label`, y después formamos las
combinaciones lineales correspondientes. Finalmente mostramos las amplitudes
numéricas de los vectores en la base computacional ordenada
$\{|000\rangle, |001\rangle, |010\rangle, |011\rangle,
  |100\rangle, |101\rangle, |110\rangle, |111\rangle\}$.


```python


# Estados base de tres qubits en Qiskit
sv_000 = Statevector.from_label('000')
sv_111 = Statevector.from_label('111')

# Combinaciones lineales correspondientes a ψ1, ψ2 y ψ3
sv_psi1 = (sv_000 + sv_111) / np.sqrt(2)
sv_psi2 = (1j * sv_000 - sv_111) / np.sqrt(2)
sv_psi3 = (sv_000 + 1j * sv_111) / np.sqrt(2)

# Mostramos las amplitudes con pocos decimales
def imprimir_estado(nombre, sv, decimales=4):
    print(nombre, "=", np.round(sv.data, decimales))

imprimir_estado("|ψ1>", sv_psi1)
imprimir_estado("|ψ2>", sv_psi2)
imprimir_estado("|ψ3>", sv_psi3)
```
